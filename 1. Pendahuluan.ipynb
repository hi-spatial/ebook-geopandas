### Contoh Penggunaan Sederhana GeoPandas

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Membaca data shapefile
gdf = gpd.read_file("./data/ne_110m_admin_0_countries/")

# Membuat visualisasi sederhana dari data geospatial
gdf.plot()
plt.title("Visualisasi Data Geospatial dengan GeoPandas")
plt.show()
```
